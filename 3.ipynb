```python
import pandas as pd
```

```python
players = pd.read_csv('data/Player.csv')
matches = pd.read_csv('data/Match.csv')
characters = pd.read_csv('data/Character.csv')
players['name'] = players['name'].replace('', pd.NA).fillna(players['discordName'])
```

```python

franz_id = players.loc[players['name'] == 'Franz', 'id'].iloc[0]
aura_id = characters.loc[characters['name'] == 'Aura', 'id'].iloc[0]

franz_aura_matches = matches[
    ((matches['winnerId'] == franz_id) & (matches['winnerCharacterId'] == aura_id)) |
    ((matches['loserId'] == franz_id) & (matches['loserCharacterId'] == aura_id))
]

franz_aura_matches.to_csv('data/franz_aura_matches.csv', index=False)
print(f"Saved {len(franz_aura_matches)} matches to data/franz_aura_matches.csv")
```
